# Exercise 1: Agriculture — Kumasi Market Crop Tracker

1. Create a list of 6 crop prices in GHS. Use slicing to print only the first 3 prices.
2. Use negative indexing to get the last price. Reverse the list with `[::-1]`. Find max, min, average.
3. Create a tuple for the market location: (name, city, lat, lng). Unpack it into 4 variables.
4. Create a dictionary for one crop: `{name, price_ghs, region, harvest_month, quantity_kg}`.
5. Use a list comprehension to get only crops with price > 20 from your list.
6. Use a list comprehension to convert all prices from GHS to USD (divide by 14.5).

```python
# 1. List of 6 crop prices in GHS
crop_prices = [12.5, 25.0, 18.2, 30.5, 22.1, 15.8]
print("First 3 prices:", crop_prices[:3])

# 2. Negative indexing, reverse, and stats
last_price = crop_prices[-1]
reversed_prices = crop_prices[::-1]
max_price = max(crop_prices)
min_price = min(crop_prices)
avg_price = sum(crop_prices) / len(crop_prices)

print(f"Last price: {last_price}")
print(f"Reversed prices: {reversed_prices}")
print(f"Max: {max_price}, Min: {min_price}, Avg: {avg_price:.2f}")

# 3. Tuple for market location and unpacking
market_location = ("Kumasi Central Market", "Kumasi", 6.6885, -1.6244)
name, city, lat, lng = market_location
print(f"Market: {name} in {city} (Lat: {lat}, Lng: {lng})")

# 4. Dictionary for one crop
crop_info = {
    "name": "Cassava",
    "price_ghs": 25.0,
    "region": "Ashanti",
    "harvest_month": "October",
    "quantity_kg": 50
}
print("Crop Info:", crop_info)

# 5. List comprehension: prices > 20
expensive_crops = [price for price in crop_prices if price > 20]
print("Prices > 20:", expensive_crops)

# 6. List comprehension: GHS to USD (divide by 14.5)
usd_prices = [round(price / 14.5, 2) for price in crop_prices]
print("Prices in USD:", usd_prices)
```

# Exercise 2: FinTech — MoMo Transaction Processor

1. Create a list of 8 transaction amounts (mix positive and negative). Find total, max, min.
2. Use a conditional (if/elif/else) to classify account status based on balance.
3. Use `if not transactions:` (truthiness check) to handle empty list.
4. Create a dict for a MoMo account. Use `.get()` and loop over `.items()`.
5. Create sets for MTN and Vodafone subscribers. Find shared subscribers (intersection).
6. Use a list comprehension to filter only deposits (positive amounts).

```python
# 1. List of 8 transactions
transactions = [500, -200, 1000, -50, 300, -400, 150, -10]
total_balance = sum(transactions)
max_trans = max(transactions)
min_trans = min(transactions)
print(f"Total: {total_balance}, Max: {max_trans}, Min: {min_trans}")

# 2. Conditional for account status
balance = total_balance
if balance < 100:
    status = "Critical"
elif balance < 500:
    status = "Low"
else:
    status = "Healthy"
print(f"Account Status: {status}")

# 3. Truthiness check for empty list
empty_list = []
if not empty_list:
    print("Warning: No transactions found!")

# 4. MoMo account dictionary and methods
momo_account = {
    "account_holder": "Julius Dagana",
    "phone_number": "0244123456",
    "network": "MTN",
    "balance": total_balance
}

# Safe access with .get()
verified = momo_account.get("is_verified", False)
print(f"Verified Status: {verified}")

# Loop over .items()
print("Account Details:")
for key, value in momo_account.items():
    print(f"- {key.replace('_', ' ').title()}: {value}")

# 5. Sets for MTN and Vodafone subscribers
mtn_subs = {"0244123456", "0544111222", "0243999888", "0555000111"}
voda_subs = {"0200111222", "0500333444", "0244123456", "0555000111"}

shared_subs = mtn_subs.intersection(voda_subs)
print("Shared Subscribers (Multi-SIM users):", shared_subs)

# 6. List comprehension for deposits
deposits = [t for t in transactions if t > 0]
print("Deposits:", deposits)
```

# Exercise 3: Healthcare — Patient Triage System

1. Create a list of 6 patient triage scores (0–100). Use `enumerate()` to print ranked list.
2. Use a conditional to classify each score (Critical, Urgent, Standard, Routine).
3. Create a tuple for each patient's fixed data (patient_id, blood_type, dob) — Why tuple?
4. Create a set of diagnoses from 8 patients (show deduplication).
5. Use `zip()` to pair names and scores. Print with `enumerate()`.
6. Use a list comprehension to get names of all Critical patients (score >= 80).

```python
# 1 & 2. Triage scores and classification
triage_scores = [85, 45, 92, 20, 65, 78]
print("Ranked Patient List:")
for i, score in enumerate(triage_scores, 1):
    if score >= 80:
        category = "Critical"
    elif score >= 60:
        category = "Urgent"
    elif score >= 40:
        category = "Standard"
    else:
        category = "Routine"
    print(f"{i}. Score: {score} -> Category: {category}")

# 3. Tuple for fixed data
# Why tuple? Because patient ID, blood type, and DOB are fixed and shouldn't be modified (immutability).
patient_data = ("P-1001", "O+", "1995-05-12")
print(f"Fixed Patient Record: {patient_data}")

# 4. Set of diagnoses (removing duplicates)
diagnoses = ["Malaria", "Flu", "Malaria", "Typhoid", "Flu", "Malaria", "Cold", "Anemia"]
unique_diagnoses = set(diagnoses)
print(f"Unique Diagnoses (Set): {unique_diagnoses}")

# 5. Pairing names and scores with zip()
patient_names = ["Kwame", "Ama", "Kofi", "Yaa", "Abena", "Yaw"]
paired_data = zip(patient_names, triage_scores)

print("\nPatient Triage Assignments:")
for i, (name, score) in enumerate(paired_data, 1):
    print(f"Rank {i}: {name} (Score: {score})")

# 6. List comprehension for Critical patients
critical_patients = [name for name, score in zip(patient_names, triage_scores) if score >= 80]
print(f"\nCritical Patients (Ghana Hospital Context): {critical_patients}")
```
