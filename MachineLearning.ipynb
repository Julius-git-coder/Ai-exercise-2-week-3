# Notebook 2: Machine Learning Foundations

Now that we have processed our startup data, we can build a simple machine learning model. 

**Goal**: Predict the total funding (`funding_usd_millions`) of a startup based on its age and number of employees.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Load the data created in the previous step
df = pd.read_csv('processed_startups.csv')

# 2. Feature Engineering: Calculate startup age
current_year = 2025
df['age'] = current_year - df['founded_year']

print("Processed Data for ML:")
display(df.head())
```

### Data Preprocessing
ML models require numeric input. We'll use One-Hot Encoding for the `sector` column.

```python
# 3. Encode categorical variables
df_encoded = pd.get_dummies(df, columns=['sector'], drop_first=True)

# Define Features (X) and Target (y)
X = df_encoded.drop(columns=['company', 'funding_usd_millions', 'founded_year'])
y = df_encoded['funding_usd_millions']

print("Features (X) Shape:", X.shape)
X.head()
```

### Model Training
> **Note**: With only 4 rows of data, the model will not be statistically significant. This is a demonstration of the workflow.

```python
# 4. Split data (Normally we would use a larger dataset)
# Using a very small test size just to show the function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 5. Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

print("Model trained successfully!")
```

### Predictions and Results

```python
# 6. Make predictions
y_pred = model.predict(X_test)

# 7. Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Check coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print("\nModel Coefficients:")
print(coefficients)
```
