```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    "City": ["Accra", "Kumasi", "Tamale", "Cape Coast"],
    "Population": [2300000, 3200000, 800000, 200000],
    "GrowthRate": [2.1, 2.5, 3.0, 1.8]
}

df = pd.DataFrame(data)

# Create figure
plt.figure()

# Bar chart
plt.bar(df["City"], df["Population"])

# Titles and labels
plt.title("Population by City")
plt.xlabel("City")
plt.ylabel("Population")

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Sales": [200, 250, 300, 280, 350]
}

df = pd.DataFrame(data)

# Line (linear) plot
plt.plot(df["Year"], df["Sales"])

# Labels and title
plt.title("Sales Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Sales")

# Show
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "City": ["Accra", "Kumasi", "Tamale", "Cape Coast"],
    "Population": [2300000, 3200000, 800000, 200000]
}

df = pd.DataFrame(data)

# Plot line with red dots
plt.plot(df["City"], df["Population"], marker='o', color='red', linestyle='-')

# Labels and title
plt.title("Population Trend")
plt.xlabel("City")
plt.ylabel("Population")

# Show grid (optional but makes it nicer)
plt.grid()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [30, 45, 38, 60, 72, 65]

plt.plot(months, sales, color='blue', marker='o')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (GHS)')
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['Kwame', 'Ama', 'Kojo', 'Abena', 'Yaw']
scores = [88, 75, 91, 70, 85]

plt.bar(students, scores, color='green')
plt.title('Student Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.show()
```
